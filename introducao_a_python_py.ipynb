#**Classes**

````python
class Minha_classe_pessoa():
    def __init__(self, nome, idade):
      self.__nome = nome
      self.__idade = idade

    def get_nome(self):
      return self.__nome

    def get_idade(self):
      return self.__idade

def main():
    lista_pessoas = []
    while True:
        opcao = int(input("1 Incluir pessoa.\n" +
                       "2 Imprimir pessoas.\n" +
                       "3 Sair.\n"))
        if opcao is 1:
            nome = input("Informe nome para inserir:\n")
            idade = input("Informe a idade de "+nome+":\n")
            pessoa = Minha_classe_pessoa(nome, idade)
            lista_pessoas.append(pessoa)
        elif opcao is 2:
            for pessoa in lista_pessoas:
              print("Nome: ", pessoa.get_nome(), "\tIdade:", str(pessoa.get_idade()))
        elif opcao is 3:
            break
        else:
            print("Opção inválida.")

if __name__ == '__main__':
    main()
````

````python
# Not mine! Adaptado de: Lucas e Felipe
import pandas as pd 

class Minha_classe:
    def __init__(self, df):
        self.df = df
        #lista das colunas do documento
        colunas = df.columns.to_list()
        self.coluna_id = colunas[0]
        self.coluna_nome = colunas[1]
        self.coluna_sobrenome = colunas[2] 
        self.coluna_email = colunas[3] 
        self.coluna_genero = colunas[4] 
        self.coluna_nfilhos = colunas[5] 
        self.coluna_datanasc = colunas[6] 
        
    def busca_sequencial(self, valor_buscado):
        index = 0
        n = len(self.df[self.coluna_id])
        cont = 0
        if valor_buscado > n:
            return print("ID inválido.")

        while index < n:
            if self.df[self.coluna_id][index] == valor_buscado:
                pos = index
                cont += 1
                return self.imprimir(pos, cont)
            else:
                index += 1
                cont += 1

    def busca_binaria(self, valor_buscado, inicio, fim, cont):
        try:
            meio = (inicio + (fim-inicio)//2)
            
            if inicio > fim:
                cont += 1
                return False
            elif self.df[self.coluna_id][meio] == valor_buscado:
                cont += 1
                return self.imprimir(meio, cont)
            elif self.df[self.coluna_id][meio] > valor_buscado:
                cont += 1
                #partir lista para inferior
                return self.busca_binaria(valor_buscado, inicio, meio-1, cont)
            else:
                #partir lista superior
                cont += 1
                return self.busca_binaria(valor_buscado, meio+1, fim, cont)
                
        except:
            return print("ID inválido.")

    #imprimir os dados
    def imprimir(self, resultado, cont):
        if resultado:
            print()
            print("ID: ", self.df[self.coluna_id][resultado])
            print("Nome: ", self.df[self.coluna_nome][resultado], self.df[self.coluna_sobrenome][resultado])
            print("Email: ", self.df[self.coluna_email][resultado])
            print("Número de filhos: ", self.df[self.coluna_nfilhos][resultado])
            print("Genero: ", self.df[self.coluna_genero][resultado])
            print("Número de comparações: ", cont) #mostrar o numero de comparações
        else:
            print("Não encontrado.")
    
#inserção dos dados do documento no código
df = pd.read_csv("Contoso_-_Clientes.csv", encoding="UTF-8", sep=";")

#lista das colunas do documento
colunas = df.columns.to_list()

#separacao das colunas
coluna_id = colunas[0]

cont = 0
                
#pede ao usuário qual busca prefere utilizar
opcao = int(input("Selecione o tipo de busca:\n" + 
                  "1- Busca Sequencial\n" + 
                  "2- Busca Binária\n" +
                  "Opção: "))

objeto = Minha_classe(df)

if opcao == 1:
    ID = int(input("Informe o ID: "))
    objeto.busca_sequencial(ID)

elif opcao == 2:
    ID = int(input("Informe o ID: "))
    n = len(df[coluna_id])
    objeto.busca_binaria(ID, 0, n, cont)
else:
    print("Opção inválida.")

df.head(ID)
````
